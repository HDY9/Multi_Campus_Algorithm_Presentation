### 안정성

- 안정성/불안정성
- 안정성: 정렬이 이뤄지기 전, 같은 값의 원소 순서가 정렬 후에도, 그대로 유지되는 것.
- 불안정성: 순서가 그대로 유지되지 않을 수 있다.

### 내부 정렬과 외부 정렬

- 내부 정렬: 정렬의 대상이 되는 모든 데이터가 하나의 배열에 저장할 수 **있음**.
- 외부 정렬: 하나의 배열에 저장할 수 **없음**.

## 버블 정렬

- 액체 속의 공기 방울이 가벼워서 위로 보글보글 올라오는 모습에 착안하여 붙인 이름, 단순 교환 정렬이라고도 함.
- 두 개의 원소(이웃한)를 단순 비교를 하는 것.
- 작은 값을 앞으로 보낸다.
- 범위 내 가장 큰 값을(or 작은 값을) 맨 뒤로(or 맨 앞으로) 보내는 작업을 반복하는 것.
- 시작점을 어디로 잡는가에 따라.
- 패스 : 비교ㆍ교환하는 과정을 지칭. (모든 정렬이 끝나려면 정렬을 n-1번 수행)

- ***세이커 정렬 (양방향 버블 정렬, 칵테일 정렬, 칵테일 셰이커 정렬)***
- 홀수 패스에서는 가장 작은 원소를 맨 앞으로 이동.
- 짝수 패스에서는 가장 큰 원소를 맨 뒤로 이동.

```python
# 실습 5
# 세이커 정렬 (양방향 버블 정렬, 칵테일 정렬, 칵테일 셰이커 정렬)

from typing import MutableSequence

def shaker_sort_verbose(a:MutableSequence) -> None:
    """세이커 정렬(정렬 과정을 출력)"""
    ccnt = 0 # 비교 횟수
    scnt = 0 # 교환 횟수
    left = 0
    n = len(a)
    right = len(a)-1
    last = right
    i = 0
    
    while left < right:
        print(f'패스{i+1}')
        i += 1
        for j in range(right,left,-1):
            for m in range(0,n-1):
                print(f'{a[m]:2}' + ('  ' if m != j-1 else
                                    ' +' if a[j-1] > a[j] else ' -'),
                     end='')
            print(f'{a[n-1]:2}')
            ccnt += 1
            if a[j-1] > a[j]:
                scnt += 1
                a[j-1],a[j] = a[j],a[j-1]
                last = j
        left = last  # 여기서 1이 저장, 두 번째 while문 돌 때 6이 저장
        for m in range(0,n-1):
            print(f'{a[m]:2}', end='  ')
        print(f'{a[n-1]:2}')
        
        if (left == right):
            break
        print(f'패스 {i+1}')
        i += 1
        for j in range(left, right):
            for m in range(0, n - 1):
                print(f'{a[m]:2}' + ('  ' if m != j else
                                    ' +' if a[j] > a[j + 1] else ' -'),
                     end='')
            print(f'{a[n - 1]:2}')
            ccnt += 1
            if a[j] > a[j + 1]:
                scnt += 1
                a[j], a[j + 1] = a[j + 1], a[j]
                last = j
        right = last  # 여기서 6이 저장
        for m in range(0, n - 1):
            print(f'{a[m]:2}', end='  ')
        print(f'{a[n - 1]:2}')
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')
    
if __name__ == '__main__':
    print('셰이커 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num  # 원소 수 num인 배열을 생성

    for i in range(num):
        x[i] = int(input(f'x[{i}] : '))

    shaker_sort_verbose(x)  # 배열 x를 단순 교환 정렬

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
        
# 9 1 3 4 6 7 8
```
